```python
#linear
#PDX
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import spearmanr
from sklearn.decomposition import PCA
import numpy as np

data = np.load("/Users/tranluan/Desktop/School/Spring 2024/Machine learning/data/pdx/data.npy", allow_pickle=True)

# Assuming the last column is the target variable and the rest are features
X = data[:, :-1]  # Features
y = data[:, -1]   # Target variable

# Initialize PCA with desired number of components
pca = PCA(n_components=100)  # Adjust the number of components as needed

# Initialize the Linear Regression model
linear_reg = LinearRegression()

# Define 3-fold cross-validation
kf = KFold(n_splits=3, shuffle=True, random_state=42)

# Number of times to repeat cross-validation
num_iterations = 10

# Initialize lists to store mean MSE and SCC values across iterations
mean_mse_values = []
mean_scc_values = []

for iteration in range(num_iterations):
    # Initialize lists to store MSE and SCC values for each fold
    mse_values = []
    scc_values = []

    # Perform 3-fold cross-validation
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Fit PCA on the training data and transform both training and test data
        X_train_pca = pca.fit_transform(X_train)
        X_test_pca = pca.transform(X_test)

        # Train the model
        linear_reg.fit(X_train_pca, y_train)

        # Predict on the test data
        y_pred = linear_reg.predict(X_test_pca)

        # Calculate MSE
        mse = mean_squared_error(y_test, y_pred)

        # Calculate SCC
        scc, _ = spearmanr(y_test, y_pred)

        # Append MSE and SCC values to lists
        mse_values.append(mse)
        scc_values.append(scc)

    # Calculate the mean MSE and SCC across folds for this iteration
    mean_mse = np.mean(mse_values)
    mean_scc = np.mean(scc_values)

    # Append mean MSE and SCC values to the lists across iterations
    mean_mse_values.append(mean_mse)
    mean_scc_values.append(mean_scc)

    # Print mean MSE and SCC for the current iteration
    print("Iteration:", iteration + 1)
    print("Mean MSE:", mean_mse)
    print("Mean SCC:", mean_scc)

# Convert lists to numpy arrays for easier calculations
mean_mse_values = np.array(mean_mse_values)
mean_scc_values = np.array(mean_scc_values)

# Calculate the overall mean and standard deviation of MSE and SCC across iterations
overall_mean_mse = np.mean(mean_mse_values)
overall_std_mse = np.std(mean_mse_values)
overall_mean_scc = np.mean(mean_scc_values)
overall_std_scc = np.std(mean_scc_values)

# Print the overall mean and range of MSE and SCC across iterations
print("\nOverall Mean MSE:", overall_mean_mse)
print("Overall MSE Range:", overall_mean_mse - overall_std_mse, "to", overall_mean_mse + overall_std_mse)
print("Overall Mean SCC:", overall_mean_scc)
print("Overall SCC Range:", overall_mean_scc - overall_std_scc, "to", overall_mean_scc + overall_std_scc)
```

```python
#PDX
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from scipy.stats import spearmanr
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeRegressor

# Load the data
data = np.load("/Users/tranluan/Desktop/School/Spring 2024/Machine learning/data/pdx/data.npy", allow_pickle=True)

# Assuming the last column is the target variable and the rest are features
X = data[:, :-1]  # Features
y = data[:, -1]   # Target variable

# Initialize PCA with desired number of components
pca = PCA(n_components=50)  # Adjust the number of components as needed

# Initialize the Decision Tree Regressor model
decision_tree = DecisionTreeRegressor()

# Define 3-fold cross-validation
kf = KFold(n_splits=3, shuffle=True, random_state=42)

# Number of times to repeat cross-validation
num_iterations = 10

# Initialize lists to store mean MSE and SCC values across iterations
mean_mse_values = []
mean_scc_values = []

for iteration in range(num_iterations):
    # Initialize lists to store MSE and SCC values for each fold
    mse_values = []
    scc_values = []

    # Perform 3-fold cross-validation
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Fit PCA on the training data and transform both training and test data
        X_train_pca = pca.fit_transform(X_train)
        X_test_pca = pca.transform(X_test)

        # Train the model
        decision_tree.fit(X_train_pca, y_train)

        # Predict on the test data
        y_pred = decision_tree.predict(X_test_pca)

        # Calculate MSE
        mse = mean_squared_error(y_test, y_pred)

        # Calculate SCC
        scc, _ = spearmanr(y_test, y_pred)

        # Append MSE and SCC values to lists
        mse_values.append(mse)
        scc_values.append(scc)

    # Calculate the mean MSE and SCC across folds for this iteration
    mean_mse = np.mean(mse_values)
    mean_scc = np.mean(scc_values)

    # Append mean MSE and SCC values to the lists across iterations
    mean_mse_values.append(mean_mse)
    mean_scc_values.append(mean_scc)

    # Print mean MSE and SCC for the current iteration
    print("Iteration:", iteration + 1)
    print("Mean MSE:", mean_mse)
    print("Mean SCC:", mean_scc)

# Convert lists to numpy arrays for easier calculations
mean_mse_values = np.array(mean_mse_values)
mean_scc_values = np.array(mean_scc_values)

# Calculate the overall mean and standard deviation of MSE and SCC across iterations
overall_mean_mse = np.mean(mean_mse_values)
overall_std_mse = np.std(mean_mse_values)
overall_mean_scc = np.mean(mean_scc_values)
overall_std_scc = np.std(mean_scc_values)

# Print the overall mean and range of MSE and SCC across iterations
print("\nOverall Mean MSE:", overall_mean_mse)
print("Overall MSE Range:", overall_mean_mse - overall_std_mse, "to", overall_mean_mse + overall_std_mse)
print("Overall Mean SCC:", overall_mean_scc)
print("Overall SCC Range:", overall_mean_scc - overall_std_scc, "to", overall_mean_scc + overall_std_scc)
```
