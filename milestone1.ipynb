```python
count = df_response_ccle['LN18'].count()
missing_percentage = (df_response_ccle['LN18'].isnull().sum() / count) * 100
cardinality = df_response_ccle['LN18'].nunique()
min_value = df_response_ccle['LN18'].min()
q1 = df_response_ccle['LN18'].quantile(0.25)
median = df_response_ccle['LN18'].median()
q3 = df_response_ccle['LN18'].quantile(0.75)
max_value = df_response_ccle['LN18'].max()
mean = df_response_ccle['LN18'].mean()
std_dev = df_response_ccle['LN18'].std()

# Create a DataFrame for the data quality report
data_quality_report = pd.DataFrame({
    'Feature': ['LN18'],
    'Desc.': ['Drug'],
    'Count': [count],
    '% of Missing': [missing_percentage],
    'Card.': [cardinality],
    'Min.': [min_value],
    'Q1': [q1],
    'Median': [median],
    'Q3': [q3],
    'Max.': [max_value],
    'Mean': [mean],
    'Std. Dev.': [std_dev],
    'Notes': ['']
})

# Display the data quality report
print(data_quality_report)
```

```python
df_new = df_response_ccle.iloc[:300, :15]
plt.figure(figsize=(10,5))
c= df_new.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
```
